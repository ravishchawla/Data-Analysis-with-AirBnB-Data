```python
def df_stats(df):
    print('Shape: ' , df.shape);
    
    missings = df.isnull().sum() / len(df);
    missing_vals = dict(zip(df.columns[missings > 0], missings[missings > 0]));
    print('# Columns with any missing elements : ' ,   [(w, missing_vals[w]) for w in sorted(missing_vals, key=missing_vals.get, reverse=True)])
    
    print();
```

```python
listings[['host_response_rate']] = listings['host_response_rate'].apply(lambda col: float(str(col).replace("%", "")))
```

```python
listings['notes_exist'] = listings['notes'].isnull().astype('int');
listings = listings.drop('notes', axis=1)
```

```python
listings['host_response_time'] = listings['host_response_time'].fillna('not provided')
```

```python
listings = listings.drop('access', axis=1)
```

```python
listings = listings.drop('transit', axis=1)
```

```python
listings_numerical = listings.select_dtypes(include=['int64', 'float64']);
```

```python
listings_categorical = listings.select_dtypes(exclude=['int64', 'float64'], );
```

```python
with pd.option_context('display.max_columns', None):
    display(listings_categorical)
```

```python
len(listings_categorical.columns)
```

```python
categorical_attributes = ['experiences_offered', 'host_location', 'host_response_time', 'host_neighbourhood', 'street', 'neighbourhood', 'neighbourhood_cleansed', 'neighbourhood_group_cleansed', 'city', 'state', 'market', 'smart_location', 'country_code', 'country', 'property_type', 'room_type', 'bed_type', 'cancellation_policy'];

len(categorical_attributes)
```

```python
full_text_attributes = ['name', 'summary', 'space', 'description', 'neighborhood_overview', 'interaction', 'house_rules', 'host_about']

len(full_text_attributes)
```

```python
dropping_attributes = ['listing_url', 'picture_url', 'host_url', 'host_thumbnail_url', 'host_picture_url', 'last_scraped', 'host_name', 'calendar_last_scraped', 'calendar_updated']

len(dropping_attributes)
```

```python
date_attributes = ['host_since', 'first_review', 'last_review']
bool_attributes = ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'is_location_exact', 'has_availability', 'requires_license', 'instant_bookable', 'is_business_travel_ready', 'require_guest_profile_picture', 'require_guest_phone_verification']
array_attributes = ['host_verifications', 'amenities', 'jurisdiction_names']
money_attributes = ['extra_people', 'cleaning_fee', 'security_deposit', 'price']
print(len(date_attributes), len(bool_attributes), len(array_attributes), len(money_attributes))
```

```python
all_attrs = set(categorical_attributes).union(full_text_attributes).union(dropping_attributes).union(date_attributes).union(bool_attributes).union(array_attributes).union(money_attributes);
all_attrs.intersection(set(categorical_attributes)) == set(categorical_attributes)
```

```python
listings_categorical_dummies = pd.get_dummies(listings_categorical[categorical_attributes]);
```

```python
listings_date_factors = listings_categorical[date_attributes].apply(lambda col: pd.factorize(col, sort=True)[0]);
```

```python
def repl_f_t(l):
    l = l.replace('f', 0);
    l = l.replace('t', 1);
    return l
    
listings_bool_replaced = listings_categorical[bool_attributes].apply(lambda col: repl_f_t(col));
```

```python
def get_arr(l):
    l = l.str.replace('[', '');
    l = l.str.replace(']', '');
    l = l.str.replace('{', '');
    l = l.str.replace('}', '');
    l = l.str.replace("'", '');
    l = l.str.replace('"', '');
    return l;
listings_arr_fixed = listings[array_attributes].apply(lambda col: get_arr(col));

array_dummy_arrs = [];
for col in listings_arr_fixed.columns:
    dummy_arrs = listings_arr_fixed[col].str.get_dummies(sep=',');
    dummy_arrs = dummy_arrs.add_prefix(col + '_');
    array_dummy_arrs.append(dummy_arrs);
    
listings_arr_fixed = pd.concat(array_dummy_arrs, axis=1);
```

```python
listings_money_floats = listings_categorical[money_attributes].apply(lambda col: col.str.replace("$", "").str.replace(",","").astype('float'))
```

```python
listings_all_cats = pd.concat([listings_categorical_dummies, listings_date_factors, listings_bool_replaced, listings_arr_fixed, listings_money_floats], axis=1)
listings_all_cats = listings_all_cats.select_dtypes(include=['float64', 'int64', 'uint8'])
```

```python
listings_all_cats[['cleaning_fee', 'security_deposit']] = listings_all_cats[['cleaning_fee', 'security_deposit']].fillna(listings_all_cats[['cleaning_fee', 'security_deposit']].mean());
```

```python
#listings_all_cats[list(listings_categorical_dummies.columns) + list(listings_arr_fixed.columns) + list(listings_bool_replaced)]
np.unique(listings_all_cats.dtypes)
```

```python
listings_all_cats[list(listings_categorical_dummies.columns) + list(listings_arr_fixed.columns) + list(listings_bool_replaced)] = listings_all_cats[list(listings_categorical_dummies.columns) + list(listings_arr_fixed.columns) + list(listings_bool_replaced)].astype('int');
```

```python
#listings_all_cats = listings_all_cats.fillna(listings_all_cats.mode().T);
for col in listings_numerical:
    listings_numerical[col] = listings_numerical[col].fillna(listings_numerical[col].mode()[0])
```

```python
print('Cat nulls: ' , listings_all_cats.isnull().sum().sum())
print('Num nulls: ' , listings_numerical.isnull().sum().sum())

print('Cat shape: ' , listings_all_cats.shape)
print('Num shape: ' , listings_numerical.shape)
```

```python
listings_cleaned = pd.concat([listings_all_cats, listings_numerical], axis=1);
```

```python
holidays_dataset = pd.read_csv('data/usholidays.csv');
holidays_dataset = holidays_dataset[['Date', 'Holiday']]
```

```python
from datetime import datetime, timedelta;
#blah = [{'Date' : '12', 'Holiday' : 'Year'}, {'Date' : '13', 'Holiday' : 'Ydear'}, {'Date' : '14', 'Holiday' : 'Yearl'}]
#pd.DataFrame(blah)

rows_list = [];
for row in holidays_dataset.iterrows():
    date = row[1]['Date'];
    date = datetime.strptime(date, '%Y-%m-%d')
    for day in range(-5, 5):
        new_date = datetime.strftime(date + timedelta(days=day), '%Y-%m-%d');
        rows_list.append({'Date' : new_date, 'Holiday' : row[1]['Holiday']})
    
holidays_dataset = pd.DataFrame(rows_list);
```

```python
calandar = calandar.merge(holidays_dataset.rename(columns={'Date' : 'date'}), how='left', on='date')
calandar['Holiday'] = calandar['Holiday'].fillna('No Holiday');
```

```python
#I want to visualize 3 things here:
# 1. Count of avalaible_f by day, and grouped by Holiday
# 2. Count of avalaible_t by day, and grouped by Holiday
# 3. Avg. of price by Holiday
```

```python
cal_holidays = calandar.groupby('date_val').first()['Holiday'];
cal_sums = calandar.groupby('date_val').sum()[['available_f', 'available_t']];

cal_sums = cal_sums.div(cal_sums.sum(axis=1), axis=0);

cal_avgs = calandar.groupby('date_val').mean()['price'];
```

```python
cal_by_day = cal_sums.merge(cal_avgs, on='date_val').merge(cal_holidays, on='date_val')
cal_by_day = cal_by_day[cal_by_day.price > 0];
```
